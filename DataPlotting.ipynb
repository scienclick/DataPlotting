## Accessing the individual columns in the data frame

```python
df.head()
```

```python
df['total_bill'].plot.kde(figsize=(16,8)) # Kernel Density Estimation plot (KDE)
```

```python
df.plot.scatter(x='total_bill',y='tip',figsize=(16,8))
```

```python
df.plot.scatter(x='total_bill',y='tip',c='size',cmap='coolwarm',figsize=(16,8))
```

# Seaborn plotting

```python
## Categorical data plots
Factorplot
boxplot
violinplot
stripplot
swarmplot
barplot
countplot
```

```python
sns.distplot(df['total_bill'])
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.distplot(df['total_bill'],kde=False, bins=30)
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.distplot(df['total_bill'],kde=True, bins=30)
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.lmplot(x='total_bill', y='tip',data=df,hue='sex')
#sns.lmplot(x='total_bill', y='tip',data=df,hue='sex',palette='coolwarm')
#sns.lmplot(x='total_bill', y='tip',data=df,hue='sex',palette='coolwarm', markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip',data=df,col='sex') 
# Using Grid: we can add more variable separation through columns and rows with the use of grid. Just indicate 
# the col or row arguments
```

```python
sns.lmplot(x='total_bill', y='tip',row='sex',col='time',data=df)
```

```python
sns.lmplot(x='total_bill', y='tip',data=df,col='day',hue='sex',aspect=0.6,size=6)
# sns.lmplot(x='total_bill', y='tip',data=df,col='day',hue='sex',palette='coolwarm',aspect=0.6,size=6)
# Aspect and size: Seaborn figures can have thier size and aspect ratio adjusted with he size and aspect parameters
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
g=sns.FacetGrid(df,col="time", row="smoker",hue="sex")
g=g.map(plt.scatter, 'total_bill','tip').add_legend()
# notice how the arguments come after plt.scatter call
```

```python
g=sns.FacetGrid(df,col="time", row="smoker",hue="sex")
g=g.map(plt.hist, 'total_bill').add_legend()
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.countplot(x='sex',data=df)
sns.set(rc={'figure.figsize':(16,8)})
```

```python
pv=pd.pivot_table(df,'total_bill','sex','smoker',aggfunc='count')
pv.plot(kind='bar')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
pv=pd.pivot_table(df,'total_bill',['sex','smoker'],['day','time'])
pv.plot(kind='bar')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.boxplot(data=df,palette='rainbow',orient='h')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.boxplot(x='day',y='tip',data=df,palette='rainbow')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.boxplot(x='day',y='tip',hue="smoker",data=df,palette='rainbow')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.boxplot(x='day',y='tip',hue="smoker",data=df,palette='coolwarm')
sns.set(rc={'figure.figsize':(16,8)})
```

```python
sns.heatmap(df.corr())
sns.set(rc={'figure.figsize':(8,8)})
```

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
sns.set(rc={'figure.figsize':(8,8)})
```

```python
sns.clustermap(df.corr(),cmap='coolwarm',standard_scale=1) # this plot shows groups the vairables by thier similarities
```
